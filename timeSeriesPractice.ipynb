```python
#More granular data
from pytz import utc

# Convert the timestamps to timezone aware with UTC timezone
start_date = pd.to_datetime('2020-01-01').tz_localize(utc)
end_date = pd.to_datetime('2020-03-31').tz_localize(utc)


filt = (df['Date'] >= start_date) & (df['Date'] <= end_date)
df.loc[filt]
```

```python
def extract_quarters(df, snapped_at):
    df['snapped_at'] = pd.to_datetime(df['snapped_at'])
    df['Quarter'] = df['snapped_at'].dt.quarter
    
    return df


df = extract_quarters(df, 'snapped_at')
df
```
