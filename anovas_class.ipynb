```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
from statsmodels.multivariate.manova import MANOVA
```

```python
# One-Way ANOVA
#testing the effect of region on charges.
model_anova_one_way = ols('charges ~ C(region)', data=df).fit()
anova_results_one_way = anova_lm(model_anova_one_way)
print(anova_results_one_way)

print("*****************")
print("*****************")
print("*****************")
# Two-Way ANOVA
#test the effect of region and smoker on charges
model_anova_two_way = ols('charges ~ C(region) + C(smoker)', data=df).fit()
anova_results_two_way = anova_lm(model_anova_two_way)
print(anova_results_two_way)

print("*****************")
print("*****************")
print("*****************")

# N-Way ANOVA
#test the effect of region, smoker, and sex on charges.
model_anova_n_way = ols('charges ~ C(region) + C(smoker) + C(sex)', data=df).fit()
anova_results_n_way = anova_lm(model_anova_n_way)
print(anova_results_n_way)
print("*****************")
print("*****************")
print("*****************")

# Two-Way ANOVA with Interaction
#test the interaction between region and smoker
model_anova_two_way_interaction = ols('charges ~ C(region) * C(smoker)', data=df).fit()
anova_results_two_way_interaction = anova_lm(model_anova_two_way_interaction)
print(anova_results_two_way_interaction)
```

```python
#ANCOVA combines ANOVA and regression to evaluate whether population means of a 
#dependent variable (e.g., charges) are equal across levels of a categorical independent 
#variable (e.g., region), while controlling for the effects of one or more continuous 
#covariates (e.g., age, bmi).

#Testing the effect of region on charges, controlling for age:
    # One-Way ANCOVA
model_ancova_one_way = ols('charges ~ C(region) + age', data=df).fit()
ancova_results_one_way = anova_lm(model_ancova_one_way)
print(ancova_results_one_way)

# Two-Way ANCOVA
#Testing the effect of region and smoker on charges, controlling for age:
model_ancova_two_way = ols('charges ~ C(region) + C(smoker) + age', data=df).fit()
ancova_results_two_way = anova_lm(model_ancova_two_way)
print(ancova_results_two_way)

# N-Way ANCOVA
#Testing the effect of region, smoker, and sex on charges, controlling for age and bmi:
model_ancova_n_way = ols('charges ~ C(region) + C(smoker) + C(sex) + age + bmi', data=df).fit()
ancova_results_n_way = anova_lm(model_ancova_n_way)
print(ancova_results_n_way)

# Two-Way ANCOVA with Interaction
#Testing the interaction effect between region and smoker on charges, controlling for age:
model_ancova_two_way_interaction = ols('charges ~ C(region) * C(smoker) + age', data=df).fit()
ancova_results_two_way_interaction = anova_lm(model_ancova_two_way_interaction)
print(ancova_results_two_way_interaction)
```

```python
#MANOVA is used when there are multiple dependent variables. In this case, let's consider charges 
#and bmi as our dependent variables and region, smoker, and sex as our independent variables.

# One-Way MANOVA
#Testing the effect of region on charges and bmi
manova_one_way = MANOVA.from_formula('charges + bmi ~ C(region)', data=df)
manova_results_one_way = manova_one_way.mv_test()
print(manova_results_one_way)

# Two-Way MANOVA
#Testing the effect of region and smoker on charges and bmi
manova_two_way = MANOVA.from_formula('charges + bmi ~ C(region) + C(smoker)', data=df)
manova_results_two_way = manova_two_way.mv_test()
print(manova_results_two_way)

# N-Way MANOVA
#Testing the effect of region, smoker, and sex on charges and bmi:
manova_n_way = MANOVA.from_formula('charges + bmi ~ C(region) + C(smoker) + C(sex)', data=df)
manova_results_n_way = manova_n_way.mv_test()
print(manova_results_n_way)

# Two-Way MANOVA with Interaction
#Testing the interaction effect between region and smoker on charges and bmi:
manova_two_way_interaction = MANOVA.from_formula('charges + bmi ~ C(region) * C(smoker)', data=df)
manova_results_two_way_interaction = manova_two_way_interaction.mv_test()
print(manova_results_two_way_interaction)
```

```python
#MANCOVA is used when there are multiple dependent variables, and you want to control 
#for the effects of one or more covariates.

# One-Way MANCOVA
#Testing the effect of region on charges and bmi, controlling for age
mancova_one_way = MANOVA.from_formula('charges + bmi ~ C(region) + age', data=df)
mancova_results_one_way = mancova_one_way.mv_test()
print(mancova_results_one_way)

# Two-Way MANCOVA
#Testing the effect of region and smoker on charges and bmi, controlling for age:
mancova_two_way = MANOVA.from_formula('charges + bmi ~ C(region) + C(smoker) + age', data=df)
mancova_results_two_way = mancova_two_way.mv_test()
print(mancova_results_two_way)

# N-Way MANCOVA
#Testing the effect of region, smoker, and sex on charges and bmi, controlling for age and children:
mancova_n_way = MANOVA.from_formula('charges + bmi ~ C(region) + C(smoker) + C(sex) + age + children', data=df)
mancova_results_n_way = mancova_n_way.mv_test()
print(mancova_results_n_way)

# Two-Way MANCOVA with Interaction
#Testing the interaction effect between region and smoker on charges and bmi, controlling for age:
mancova_two_way_interaction = MANOVA.from_formula('charges + bmi ~ C(region) * C(smoker) + age', data=df)
mancova_results_two_way_interaction = mancova_two_way_interaction.mv_test()
print(mancova_results_two_way_interaction)
```

```python
# #Interpreting the results
# ANOVA: The p-value for the Group variable is greater than 0.05, 
#indicating that there is no statistically significant difference in Y1 across the different groups.

# ANCOVA: The p-value for the Group variable less than 0.05, indicating that there 
#is a statistically significant difference in Y1 across the different groups when controlling for Covariate. 
#The covariate itself is highly significant (p < 0.001).

# MANOVA: The test indicates significant differences across the groups for the combination of Y1 and Y2 
#(all p-values < 0.001).

# MANCOVA: The test indicates significant differences across the groups for the combination of Y1 and Y2 
#when controlling for Covariate (all p-values < 0.001). The covariate also has a significant effect 
#on the combination of Y1 and Y2 (p < 0.001).
```
