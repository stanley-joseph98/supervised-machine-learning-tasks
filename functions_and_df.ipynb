```python
import pandas as pd

def fill_missing_dates(df, date_col='date', time_interval_col='Time_Interval'):
    df[date_col] = pd.to_datetime(df[date_col])

    min_date = df[date_col].min()
    max_date = df[date_col].max()

    complete_date_range = pd.date_range(min_date, max_date, freq='D')

    df = df.set_index(date_col).reindex(complete_date_range, fill_value=None)

    # Create a new DatetimeIndex with the desired frequency
    df.index = pd.date_range(min_date, periods=len(df), freq='D')  

    df[time_interval_col] = (df.index - df.index.shift(1)) * -1

    return df

# Apply the function
df1 = fill_missing_dates(df1, 'date', 'Time_Interval')
```
